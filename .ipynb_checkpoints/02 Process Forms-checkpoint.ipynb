```python
import pandas as pd
import urllib.request, json 
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup
import PyPDF2
import os
from os import walk
import os.path
from os import path
import numpy as np
#!pip install py-readability-metrics
from readability import Readability
#!python -m nltk.downloader punkt

import time
from datetime import date
```

```python
def reCase(text):
    output = re.sub("(\w|\d)(_|-)(\w|\d)","\\1 \\3",text)
    output = re.sub("([a-z])([A-Z]|\d)","\\1 \\2",output)
    output = re.sub("(\d)([A-Z]|[a-z])","\\1 \\2",output)
    return output
```

```python
def read_pdf (file):
    try:
        pdfFile = PyPDF2.PdfFileReader(open(file, "rb"))
        if pdfFile.isEncrypted:
            try:
                pdfFile.decrypt('')
                #print ('File Decrypted (PyPDF2)')
            except:
                #
                #
                # This didn't go so well on my Windows box so I just ran this in the pdf folder's cmd:
                # for %f in (*.*) do copy %f temp.pdf /Y && "C:\Program Files (x86)\qpdf-8.0.2\bin\qpdf.exe" --password="" --decrypt temp.pdf %f
                #
                #
                #
                
                command="cp "+file+" tmp/temp.pdf; qpdf --password='' --decrypt tmp/temp.pdf "+file
                os.system(command)
                #print ('File Decrypted (qpdf)')
                #re-open the decrypted file
                pdfFile = PyPDF2.PdfFileReader(open(file, "rb"))
        text = ""
        for page in pdfFile.pages:
            text = text + " " + page.extractText()
        return text
    except:
        return ""
```

```python
def parse_form(fileloc,title=None,cat=None):
    f = PyPDF2.PdfFileReader(fileloc)
    npages = f.getNumPages()
    ff = f.getFields()
    if ff:
        fields = list(ff.keys())
    else:
        fields = []
    f_per_page = len(fields)/npages
    text = read_pdf(fileloc)
    readbility = Readability(text).flesch_kincaid().grade_level
    
    if title is None:
        title = reCase(re.search("(.*)\n",text).group(1).strip())
    
    stats = {
            "title":title,
            "category":cat,
            "pages":npages,
            "reading grade level": int(readbility),
            "avg fields per page": f_per_page,
            "fields":fields
            }    
    return stats
```

```python
# Form parsing should work by iterating over the df rows and pulling from location

def pars_form(jur,cat,source,path):
    global files_df
    
    today = date.today().strftime("%Y-%m-%d")
    
    f = []
    for (dirpath, dirnames, filenames) in walk(path):
        f.extend(filenames)
        break
    f1 = [file for file in f]

    for file in f1:
        print("Opening: "+file)
        #print(path+file)
        #try:
        stats = parse_form(path+file,None,cat)
        title = stats["title"]
        pages = stats["pages"]
        fields = stats["fields"]
        f_per_p = stats["avg fields per page"]
        reading = stats["reading grade level"]
        url = "unknown"
        files_df = files_df.append(pd.DataFrame([[hash(title+file+jur),title,cat,file,pages,fields,f_per_p,reading,today,jur,source,url]],columns=["id","title","category","filename","pages","fields","f_per_p","reading","downloaded","jurisdiction","soruce","url"]), ignore_index=True,sort=False)
        #except:
        #    print("parsing error!")

    display(files_df.head())
```

```python
jur = "MI"
cat = "Civil Infractions"
source = "https://www.courts.michigan.gov/SCAO-forms/"
path = "data/raw/www.courts.michigan.gov/SCAO-forms/civil-infraction/"

pars_form(jur,cat,source,path)
```
